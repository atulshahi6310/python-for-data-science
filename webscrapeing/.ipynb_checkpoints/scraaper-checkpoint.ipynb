```python
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime

def scrape_books():
    # Loop through pages 1–10
    for i in range(1, 11):
        url = f"https://books.toscrape.com/catalogue/page-{i}.html"
        print(f"Scraping page {i}...")

        res = requests.get(url)
        soup = BeautifulSoup(res.text, "html.parser")
        books = soup.find_all("article", class_="product_pod")

        # Create a separate file for each page
        filename = f"page{i}.txt"   # <-- dynamic file name
        with open(filename, "w", encoding="utf-8") as f:
            f.write(f"Scrape run at: {datetime.now()}\n")
            f.write("="*100 + "\n")

            for book in books:
                book_name = book.h3.a["title"]
                book_price = book.find("p", class_="price_color").text[1:]
                f.write(f"{book_name} → {book_price}\n")

            f.write("="*100 + "\n")

if __name__ == "__main__":
    while True:
        scrape_books()
        print("Scraping done ✅, waiting 5 minutes before next run...")
        time.sleep(3)  # sleep 5 minutes
```
