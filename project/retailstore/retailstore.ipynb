### üìù Problem Statement (Day 10):

A retail store wants to analyze its **sales data** for the past 20 days.
You are tasked to:

1. **Load & explore the dataset** using **Pandas**.
2. Find the **total sales per product category**.
3. Identify the **day with maximum revenue**.
4. Calculate the **average order value**.
5. Create **visualizations** using Matplotlib & Seaborn:

   * Bar chart ‚Üí Sales per category
   * Line plot ‚Üí Daily revenue trend
   * Heatmap ‚Üí Correlation between quantity, price & revenue

üìä **Objective**: Help the store understand **which categories perform best** and **when peak sales occur**.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python


df = pd.read_csv("retail_sales_day10.csv")


print(df.head())




total_revenue = df["Revenue"].sum()
print("Total Revenue:", total_revenue)


top_products = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Products by Revenue:\n", top_products)

category_revenue = df.groupby("Category")["Revenue"].sum()
print("\nRevenue by Category:\n", category_revenue)
```

```python


plt.figure(figsize=(8,5))
sns.barplot(x=category_revenue.index, y=category_revenue.values, palette="viridis")
plt.title("Revenue by Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()

```

```python

plt.figure(figsize=(8,5))
sns.barplot(x=top_products.index, y=top_products.values, palette="magma")
plt.title("Top 5 Products by Revenue")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()


```

```python
daily_revenue = df.groupby("Date")["Revenue"].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x=daily_revenue.index, y=daily_revenue.values, marker="o")
plt.title("Daily Revenue Trend")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```
