### üìù Problem Statement (Day 11: Fitness Tracker Data) üèÉ‚Äç‚ôÇÔ∏èüí™

A fitness app records **daily activity data** for its users. You‚Äôve been given a dataset of **15 days** with details such as:

* Date
* Steps Taken
* Calories Burned
* Workout Duration (minutes)
* Sleep Hours

Your task is to:

1. Load the dataset using **Pandas**.
2. Find the **day with maximum steps** and **day with minimum sleep**.
3. Calculate **average steps, calories, and sleep hours**.
4. Identify correlation between **steps vs calories** and **sleep vs workout duration**.
5. Visualize insights using **Matplotlib & Seaborn**:

   * Line plot ‚Üí Steps & Calories over time
   * Bar chart ‚Üí Sleep Hours per day
   * Scatter plot ‚Üí Steps vs Calories

üìä **Objective**: Provide insights into **daily activity balance** and how lifestyle habits impact fitness.

---



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json


```

```python

with open("fitness_day11.json", "r") as f:
    data = json.load(f)




```

```python
df = pd.DataFrame(data)
```

```python
print("Average Steps:", df["Steps"].mean())
print("Average Calories:", df["Calories"].mean())
print("Average Sleep Hours:", df["Sleep_Hours"].mean())
print("Day with Max Steps:", df.loc[df["Steps"].idxmax(), "Date"])
print("Day with Min Sleep:", df.loc[df["Sleep_Hours"].idxmin(), "Date"])
```

```python
print("\nCorrelation Matrix:\n", df.corr(numeric_only=True))
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.lineplot(x="Date", y="Steps", data=df, marker="o", label="Steps")
sns.lineplot(x="Date", y="Calories", data=df, marker="o", label="Calories")
plt.xticks(rotation=45)
plt.title("Steps & Calories Over Time")
```

```python
plt.subplot(1,2,2)
sns.scatterplot(x="Steps", y="Calories", data=df, hue="Workout_Min", palette="viridis", s=100)
plt.title("Steps vs Calories (Workout Influence)")

plt.tight_layout()
plt.show()
```
