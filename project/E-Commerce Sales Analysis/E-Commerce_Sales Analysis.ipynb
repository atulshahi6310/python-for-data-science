### **Day 15 â€“ E-Commerce Sales Analysis** ðŸ›’ðŸ’³

**Problem:**
You are given a dataset of an online storeâ€™s sales with:

* `Order ID`
* `Product Category` (Electronics, Fashion, Home, Books, etc.)
* `Quantity`
* `Price`
* `Order Date`
* `Payment Method` (Credit Card, UPI, COD, Wallet)

**Tasks:**

1. Calculate the **total revenue** generated.
2. Find the **best-selling product category**.
3. Find the **most used payment method**.
4. Identify the **day with highest sales revenue**.
5. Visualize:

   * Bar chart â†’ Revenue by product category
   * Pie chart â†’ Payment method distribution
   * Line chart â†’ Daily revenue trend

ðŸ“Š **Objective**: This mirrors **real-world business analytics** where e-commerce platforms use data to optimize sales & marketing strategies.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = [
    {"Order ID": 1, "Product Category": "Electronics", "Quantity": 2, "Price": 15000, "Order Date": "2025-08-01", "Payment Method": "UPI"},
    {"Order ID": 2, "Product Category": "Fashion", "Quantity": 3, "Price": 2000, "Order Date": "2025-08-01", "Payment Method": "Credit Card"},
    {"Order ID": 3, "Product Category": "Home", "Quantity": 1, "Price": 5000, "Order Date": "2025-08-02", "Payment Method": "COD"},
    {"Order ID": 4, "Product Category": "Books", "Quantity": 5, "Price": 500, "Order Date": "2025-08-02", "Payment Method": "Wallet"},
    {"Order ID": 5, "Product Category": "Electronics", "Quantity": 1, "Price": 25000, "Order Date": "2025-08-03", "Payment Method": "UPI"},
    {"Order ID": 6, "Product Category": "Fashion", "Quantity": 4, "Price": 1500, "Order Date": "2025-08-03", "Payment Method": "COD"},
    {"Order ID": 7, "Product Category": "Home", "Quantity": 2, "Price": 8000, "Order Date": "2025-08-04", "Payment Method": "Credit Card"},
    {"Order ID": 8, "Product Category": "Books", "Quantity": 10, "Price": 300, "Order Date": "2025-08-04", "Payment Method": "Wallet"},
    {"Order ID": 9, "Product Category": "Electronics", "Quantity": 1, "Price": 12000, "Order Date": "2025-08-05", "Payment Method": "Credit Card"},
    {"Order ID": 10, "Product Category": "Fashion", "Quantity": 2, "Price": 2500, "Order Date": "2025-08-05", "Payment Method": "UPI"},
]
```

```python
df = pd.DataFrame(data)
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]

print("Sample Dataset:\n", df.head())
```

```python
total_revenue = df["Revenue"].sum()
print("\nTotal Revenue:", total_revenue)
```

```python
best_category = df.groupby("Product Category")["Revenue"].sum().idxmax()
print("\nBest-Selling Category:", best_category)
```

```python
top_payment = df["Payment Method"].value_counts().idxmax()
print("\nMost Used Payment Method:", top_payment)
```

```python
top_day = df.groupby("Order Date")["Revenue"].sum().idxmax()
print("\nDay with Highest Sales Revenue:", top_day.date())
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Product Category", y="Revenue", data=df, estimator=sum, ci=None, palette="viridis")
plt.title("Revenue by Product Category")
plt.ylabel("Total Revenue")
plt.show()
```

```python

df["Payment Method"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, colors=sns.color_palette("pastel"))
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()

```

```python
plt.figure(figsize=(10,5))
daily_revenue = df.groupby("Order Date")["Revenue"].sum().reset_index()
sns.lineplot(x="Order Date", y="Revenue", data=daily_revenue, marker="o")
plt.title("Daily Revenue Trend")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```
