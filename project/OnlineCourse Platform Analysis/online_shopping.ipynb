**Problem:**
You are given a dataset of students using an online learning platform. It contains:

* `Student ID`
* `Course Name`
* `Hours Spent`
* `Quiz Scores`
* `Completion Status` (Completed / In Progress / Dropped)

**Tasks:**

1. Find the **average time spent per course**.
2. Analyze the **average quiz scores per course**.
3. Count how many students **completed vs dropped vs in progress**.
4. Visualize:

   * Bar chart â†’ Average time & quiz score per course
   * Pie chart â†’ Completion status distribution



```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
np.random.seed(42)

```

```python

student_ids = range(1, 31)  # 30 students
courses = np.random.choice(["Python", "Data Science", "Machine Learning", "AI", "Web Development"], size=30)
hours_spent = np.random.randint(5, 50, size=30)
quiz_scores = np.random.randint(40, 100, size=30)
completion_status = np.random.choice(["Completed", "In Progress", "Dropped"], size=30, p=[0.5, 0.3, 0.2])

```

```python

course_df = pd.DataFrame({
    "Student ID": student_ids,
    "Course Name": courses,
    "Hours Spent": hours_spent,
    "Quiz Scores": quiz_scores,
    "Completion Status": completion_status
})


avg_time = course_df.groupby("Course Name")["Hours Spent"].mean()

```

```python

avg_scores = course_df.groupby("Course Name")["Quiz Scores"].mean()

```

```python

status_count = course_df["Completion Status"].value_counts()
```

```python
plt.figure(figsize=(14, 5))


plt.subplot(1, 2, 1)
avg_df = pd.DataFrame({"Avg Hours Spent": avg_time, "Avg Quiz Scores": avg_scores})
avg_df.plot(kind="bar", ax=plt.gca())
plt.title("Average Hours & Quiz Scores per Course")
plt.ylabel("Average Value")

plt.subplot(1, 2, 2)
status_count.plot.pie(autopct="%1.1f%%", startangle=90, colors=sns.color_palette("pastel"))
plt.title("Completion Status Distribution")
plt.ylabel("")

plt.tight_layout()
plt.show()

course_df.head()
```
