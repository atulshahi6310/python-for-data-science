You are given data from a fitness tracker app recording daily steps, calories burned, active minutes, and heart rate.

Your tasks:

Calculate average daily steps and check if the user meets the 10,000 steps goal.

Find the day with maximum calories burned üî•.

Plot:

A line graph of steps over time.

A bar chart of calories burned per day.

A scatter plot of steps vs calories.

Create a correlation heatmap (Steps, Calories, Heart Rate, Active Minutes).

üìä Skills Used: Python, Pandas, NumPy, Matplotlib, Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("fitness_day9.csv")


avg_steps = df["Steps"].mean()
```

```python
max_calories_day = df.loc[df["Calories Burned"].idxmax()]
plt.plot(df["Date"], df["Steps"], marker='o', linestyle='-', label="Steps")
plt.axhline(10000, color="red", linestyle="--", label="Goal: 10,000 steps")
plt.xticks(rotation=45)
plt.title("Daily Steps Over Time")
plt.ylabel("Steps")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Bar chart: Calories burned
plt.figure(figsize=(10, 5))
plt.bar(df["Date"], df["Calories Burned"], color="orange")
plt.xticks(rotation=45)
plt.title("Calories Burned Per Day")
plt.ylabel("Calories")
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(x="Steps", y="Calories Burned", data=df, s=100, color="blue")
plt.title("Steps vs Calories Burned")
plt.xlabel("Steps")
plt.ylabel("Calories Burned")
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(7, 5))
sns.heatmap(df[["Steps", "Calories Burned", "Active Minutes", "Avg Heart Rate (bpm)"]].corr(), 
            annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

(avg_steps, max_calories_day)
```
